```python
import numpy as np
import pandas as pd
```

```python


# Load your dataset
file_path = r"D:\Tech_Hack\TechSprint_CodeRunners\ai-ml\Data\kathmandu_emergency_data.csv"
df = pd.read_csv(file_path)

# Clean the Category column: remove spaces and lowercase
df['Category_clean'] = df['Category'].astype(str).str.strip().str.lower()

# Filter only hospitals
hospital_df = df[df['Category_clean'] == 'hospital']

# Select the required columns
columns_needed = ['Name', 'Phone', 'Address', 'Latitude', 'Longitude', 'Website', 'Place ID']
hospital_df = hospital_df[columns_needed]

# Reset index for cleanliness
hospital_df = hospital_df.reset_index(drop=True)

# Display all hospital rows in readable format
for idx, row in hospital_df.iterrows():
    print(f"Hospital {idx + 1}:")
    print(f"Name: {row['Name']}")
    print(f"Phone: {row['Phone']}")
    print(f"Address: {row['Address']}")
    print(f"Latitude: {row['Latitude']}")
    print(f"Longitude: {row['Longitude']}")
    print(f"Website: {row['Website']}")
    print(f"Place ID: {row['Place ID']}")
    print("-" * 50)

# Save filtered hospital data to new CSV
hospital_df.to_csv(r"D:\Tech_Hack\TechSprint_CodeRunners\ai-ml\Data\hospital_data.csv", index=False)
print("✅ Hospital data saved to 'hospital_data.csv'")
```
